# while loop:

# Print Numbers 1 to 5

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

# Countdown from 10

```python
i = 10
while i >= 1:
    print(i)
    i -= 1
```

# Print “Hello” Until User Types “stop”

```python
text = ""
while text != "stop":
    text = input("Type 'stop' to exit: ")
```

# Add Numbers Until User Enters 0

```python
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter another number (0 to stop): "))
print("Total:", total)
```
