# Write a program to check if a number is prime.

```python
# Take a number as input
num = int(input("Enter a number: "))

# Check if the number is prime
if num > 1:
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a Prime number")
    else:
        print(num, "is not a Prime number")
else:
    print(num, "is not a Prime number")
```

# Calculate the GCD using loops and if-else statements.

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Find GCD using loop
gcd = 1

# Find the smaller number
if num1 < num2:
    smaller = num1
else:
    smaller = num2

# Check divisors
for i in range(1, smaller + 1):
    if num1 % i == 0 and num2 % i == 0:
        gcd = i

# Print the result
print("GCD of", num1, "and", num2, "is:", gcd)
```

# Reverse a given integer without converting it to a string.

```python
# Take an integer input
num = int(input("Enter an integer: "))

# Handle negative numbers
negative = False
if num < 0:
    negative = True
    num = -num

# Reverse the integer without converting to string
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10

# Restore sign if negative
if negative:
    reversed_num = -reversed_num

# Print result
print("Reversed number:", reversed_num)
```

# Check whether a number is an Armstrong number.

```python
# Take input
num = int(input("Enter a number: "))

# Store original number
temp = num

# Count number of digits
digits = 0
t = num
while t > 0:
    digits += 1
    t //= 10

# Calculate Armstrong sum
armstrong_sum = 0
t = num

while t > 0:
    digit = t % 10
    armstrong_sum += digit ** digits
    t //= 10

# Check condition
if armstrong_sum == temp:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

# Count the frequency of each character in a string.

```python
# Take string input
text = input("Enter a string: ")

# Dictionary to store frequency
freq = {}

# Count frequency of each character
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

# Print result
print("Character frequencies:")
for char, count in freq.items():
    print(char, ":", count)
```

# Build a basic calculator that takes two numbers and an operator and performs the operation.

```python
# Take input from user
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

# Perform calculation using if-elif-else
if operator == "+":
    result = num1 + num2
    print("Result:", result)

elif operator == "-":
    result = num1 - num2
    print("Result:", result)

elif operator == "*":
    result = num1 * num2
    print("Result:", result)

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero is not allowed")

else:
    print("Invalid operator")
```

# Take a number and calculate the sum of the factorials of its digits

```python
num = int(input("Enter a number: "))

total = 0
temp = num

while temp > 0:
    digit = temp % 10

    # Find factorial of digit
    fact = 1
    for i in range(1, digit + 1):
        fact *= i

    total += fact
    temp //= 10

print("Sum of factorials of digits =", total)
```

# Take an integer and count how many digits are even and how many are odd.

```python
num = int(input("Enter a number: "))

even_count = 0
odd_count = 0

while num > 0:
    digit = num % 10

    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

    num //= 10

print("Even digits:", even_count)
print("Odd digits:", odd_count)
```

# Take a number and sum all digits that are prime (2, 3, 5, 7).

```python
num = int(input("Enter a number: "))

sum_prime_digits = 0

while num > 0:
    digit = num % 10

    if digit in [2, 3, 5, 7]:
        sum_prime_digits += digit

    num //= 10

print("Sum of prime digits =", sum_prime_digits)
```

# Print a diamond-shaped star pattern for a given odd number N.

```python
n = int(input("Enter an odd number: "))

# Upper part
for i in range(1, n + 1, 2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)

# Lower part
for i in range(n - 2, 0, -2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)
```
