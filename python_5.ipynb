# Object-Oriented Programming (OOP) Concepts in Python

```python
# Example

class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

my_car = Car("Toyota", "Red")
print(my_car.brand)
```

# Creating Classes and Objects

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def greet(self):
        print("Hello,", self.name)

s1 = Student("Amit", 90)
s1.greet()
```

# Inheritance

```python
class Parent:
    def display(self):
        print("This is Parent")

class Child(Parent):
    def show(self):
        print("This is Child")

obj = Child()
obj.display()
obj.show()
```

# Polymorphism

```python
class Bird:
    def sound(self):
        print("Chirp")

class Dog:
    def sound(self):
        print("Bark")

for animal in (Bird(), Dog()):
    animal.sound()
```

# Encapsulation

```python
class BankAccount:
    def __init__(self):
        self.__balance = 1000

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance 
```

# Abstraction

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def area(self):
        print("Area = πr²")
obj = Circle()
obj.area()
```

# Decorators

```python
def greet_decorator(func):
    def wrapper():
        print("Hello!")
        func()
        print("Have a great day!")
    return wrapper

@greet_decorator
def say_name():
    print("I’m Python")

say_name()
```

# Café Bill Calculator Input the number of coffees, sandwiches, and pastries bought. Compute the total bill using fixed prices.


```python
# Café Bill Calculator

coffee_price = 120
sandwich_price = 80
pastry_price = 60

# Input from user
coffees = int(input("Enter number of coffees: "))
sandwiches = int(input("Enter number of sandwiches: "))
pastries = int(input("Enter number of pastries: "))

# Calculate total bill
total_bill = (coffees * coffee_price) + (sandwiches * sandwich_price) + (pastries * pastry_price)

# Display bill
print("Total Bill = ₹", total_bill)
```

# Temperature Converter Take Celsius input and convert to Fahrenheit and Kelvin.

```python
# Temperature Converter

# Input temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# Convert to Fahrenheit and Kelvin
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

# Display results
print("Temperature in Fahrenheit:", fahrenheit)
print("Temperature in Kelvin:", kelvin)
```

# Attendance Warning System Input total classes and attended classes. Print attendance percentage and whether the student is eligible for exams.

```python
# Attendance Warning System

# Input from user
total_classes = int(input("Enter total number of classes: "))
attended_classes = int(input("Enter number of attended classes: "))

# Calculate attendance percentage
attendance_percentage = (attended_classes / total_classes) * 100

# Display attendance percentage
print("Attendance Percentage =", attendance_percentage, "%")

# Check eligibility
if attendance_percentage >= 75:
    print("Student is eligible for exams")
else:
    print("Student is NOT eligible for exams")
```

# Word Reverser Take a sentence and print the words in reverse order using string splitting and joining.


```python
# Word Reverser

# Input from user
sentence = input("Enter a sentence: ")

# Split sentence into words
words = sentence.split()

# Reverse words and join them
reversed_sentence = " ".join(words[::-1])

# Display result
print("Reversed sentence:", reversed_sentence)
```

# Simple Interest Calculator Input principal, rate, and time. Compute simple interest.

# Find the Largest of Three Numbers Using if-elif-else, print the largest among three numbers.

```python
# Find the Largest of Three Numbers

# Input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Check the largest number
if num1 >= num2 and num1 >= num3:
    print("Largest number is:", num1)

elif num2 >= num1 and num2 >= num3:
    print("Largest number is:", num2)

else:
    print("Largest number is:", num3)
```

# Leap Year Checker Input a year and check whether it’s a leap year.

```python
# Leap Year Checker

# Input from user
year = int(input("Enter a year: "))

# Check leap year
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

# Shopping Discount System If the purchase amount > 5000, give 10% discount; else, print “No Discount”.

```python
# Shopping Discount System

# Input from user
amount = float(input("Enter purchase amount: "))

# Check discount eligibility
if amount > 5000:
    discount = amount * 0.10
    final_amount = amount - discount

    print("Discount =", discount)
    print("Final Amount =", final_amount)

else:
    print("No Discount")
```

# Grade Generator Input marks and print grade as per score using conditions (A, B, C, D, Fail).

```python
# Grade Generator

# Input marks from user
marks = float(input("Enter marks: "))

# Check grade using conditions
if marks >= 90:
    print("Grade: A")

elif marks >= 75:
    print("Grade: B")

elif marks >= 60:
    print("Grade: C")

elif marks >= 40:
    print("Grade: D")

else:
    print("Grade: Fail")
```

# Name Shortener App Input a full name and output initials (e.g., Ravi Kishan Patel → R.K.P.).

```python
# Name Shortener App

# Input full name
full_name = input("Enter your full name: ")

# Split name into words
names = full_name.split()

# Get initials
initials = ""

for name in names:
    initials += name[0].upper() + "."

# Display result
print("Initials:", initials)
```
